```python
# Функция для определения количества и доли пропусков
def show_missing_stats(tmp0):
    """
    Функция для отображения статистики пропущенных значений в DataFrame.
    """
    missing_stats = pd.DataFrame({
        'Кол-во пропусков': tmp0.isnull().sum(),
        'Доля пропусков': tmp0.isnull().mean()
    })
    missing_stats = missing_stats[missing_stats['Кол-во пропусков'] > 0]
    
    if missing_stats.empty:
        return "Пропусков в данных нет"
    
    # Форматируем при выводе через Styler
    return (missing_stats.style.format({'Доля пропусков': '{:.4f}'}).background_gradient(cmap='coolwarm'))
```

### 1.2. Предобработка данных

Сохраним исходные датасеты и создадим копии наших датафреймов. Дальнейшие предобработку и преобразования будем проводить в копиях

```python
# Создаём копию investments
invest_df = investments.copy()

# Создаём копию returns
returns_df = returns.copy()
```

Проверьте названия столбцов в датасетах: все ли они точно отражают содержимое данных и оформлены в удобном для работы стиле. При необходимости приведите их к единому аккуратному стилю.

Проверим названия столбцов датасетов

Название столбцов датасетов написаны в стиле snake_case, однако в датасете `investments` некоторые столбцы  содержат лишние пробелы. Скорректируем название этих столбцов.

```python
# Напишем функцию для удаления лишних пробелов в названиях столбцов
def remove_spaces(df):
    """
    Удаляет все пробелы из названий столбцов.
    """
    df.columns = df.columns.str.replace(' ', '', regex=False)
    return df

# Применяем функцию к нашим датасетам
invest_df = remove_spaces(invest_df)
returns = remove_spaces(returns)

# Проверим результат
print('Названия столбцов в invest_df:')
print(invest_df.columns)
print('\nНазвания столбцов в returns:')
print(returns.columns)
```

Уберите в столбце `funding_total_usd` выделение разрядов и приведите его к числовому типу.

```python
# Напишем функцию, которая убирает выделение разрядов и преобразует тип данных в числовой
def clean_to_numeric(series):
    # Удаляем все нецифровые символы, кроме точки
    cleaned_str = series.str.replace(r'[^\d.]', '', regex=True)
    # Преобразуем в число (NaN для некорректных значений)
    numeric = pd.to_numeric(cleaned_str, downcast='float', errors='coerce')
    return numeric

# Применим функцию к столбцу funding_total_usd
invest_df['funding_total_usd'] = clean_to_numeric(invest_df['funding_total_usd'])

# Проверим результат
print('Тип данных:')
print(invest_df['funding_total_usd'].dtypes)
print('\nПервые 5 значений:')
print(invest_df['funding_total_usd'].head())
```

Обработайте типы данных в столбцах, которые хранят значения даты и времени, если это необходимо.

Преобразуем типы данных столбцов, содержащих даты к типу `datetime64`

```python
# Преобразуем типы данных столбцов, содержащих даты
for column in ['founded_at', 'first_funding_at', 'mid_funding_at', 'last_funding_at']:
    invest_df[column] = pd.to_datetime(invest_df[column],
                                         errors='coerce',
                                         format='%Y.%m.%d').astype('datetime64')
    
# Проверим результат
print('Тип данных:')
print(invest_df[['founded_at', 'first_funding_at', 'mid_funding_at', 'last_funding_at']].dtypes)
print('\nПервые 5 значений:')
print(invest_df[['founded_at', 'first_funding_at', 'mid_funding_at', 'last_funding_at']].head())
```

В датасете `cb_returns` сделайте столбец `year` индексом всего датасета, если не делали это при загрузке.

```python
# Зададим столбец year индексом датасета returns_df
returns_df = returns_df.set_index('year')

# Проверим результат
print(returns_df.index.name)
```

Обработайте текстовые данные, если это необходимо. Пропуски в текстовых столбцах заполните заглушками там, где это понадобится.

```python
# Заменяем пропуски на заглушки в текстовых столбцах
for col in ['name', 'category_list', 'market', 'status', 'country_code', 'state_code', 'region', 'city']:
    invest_df[col] = invest_df[col].astype(str).str.lower().str.strip()
    invest_df[col] = invest_df[col].replace('nan', 'unknown')

# Проверим результат
display(invest_df[['name', 'category_list', 'market', 'status', 'country_code', 'state_code', 'region', 'city']].head())
```

Заполним пропуски столбцов `founded_month` и `founded_quarter` значениями из столбца `founded_at`

```python
# Заполним пропуски столбца founded_month
# Создаём временную серию с форматом YYYY-MM из founded_at
temp_month = invest_df['founded_at'].dt.strftime('%Y-%m')

# Заполняем пропуски в founded_month значениями из temp_month
invest_df['founded_month'] = invest_df['founded_month'].fillna(temp_month)

# Проверим результат
print('Тип данных:')
print(invest_df['founded_month'].dtypes)
print('\nПервые 5 значений:')
print(invest_df['founded_month'].head())
```

```python
# Заполним пропуски столбца founded_quarter
# Создаём временную серию с форматом YYYY-QN из founded_at
temp_quarter = (
    invest_df['founded_at'].dt.strftime('%Y') +
    '-Q' +
    invest_df['founded_at'].dt.quarter.astype(str)
)

# Убираем лишние символы
temp_quarter = temp_quarter.str[:-2]

# Заполняем пропуски в founded_quarter значениями из temp_quarter
invest_df['founded_quarter'] = invest_df['founded_quarter'].fillna(temp_quarter)

# Проверим результат
print('Тип данных:')
print(invest_df['founded_quarter'].dtypes)
print('\nПервые 5 значений:')
print(invest_df['founded_quarter'].head())
```

Обработайте полные дубликаты в данных и пропуски в `funding_total_usd`. избавьтесь от тех строк, которые не несут какой-либо информации либо не содержат данных о финансировании.

Проверим наличие полных дубликатов

Проверим наличие неявных дубликатов. Сделаем это по столбцам `name`, `founded_at` и `funding_total_usd`. Сочетание дубликатов в этих трёх столбцах говорит о том, что записи являются неявными дубликатами.

```python
print('Количество строк до удаления неявных дубликатов:')
print(invest_df.shape[0])

# Выделяем неявные дубликаты
duplicates = invest_df[invest_df.duplicated(subset=['name', 'founded_at', 'funding_total_usd'], keep=False)]
print(duplicates)

# Удаляем неявные дубликаты
invest_df = invest_df.drop_duplicates(subset=['name', 'founded_at', 'funding_total_usd'])

# Контролируем количество строк
print('\nКоличество строк после удаления неявных дубликатов:')
print(invest_df.shape[0])
```

Нашли и удалили 4855 полных и 6 неявных дубликатов. Таким образом мы сможем избежать искажений в анализе.

Переходим к обработке пропусков в `funding_total_usd` и избавимся от тех строк, которые не несут какой-либо информации либо не содержат данных о финансировании.

```python
# Удаляем строки, где нет данных об общем объёме финансирования
invest_df = invest_df.dropna(subset=['funding_total_usd'])

# Избавляемся от строк, где финансирование равно 0
invest_df = invest_df[invest_df['funding_total_usd'] > 0]

# Удаляем строки, где нет названия компании
invest_df = invest_df[invest_df['name'] != 'unknown']

# Контролируем количество строк
print('\nКоличество строк после очистки:')
print(invest_df.shape[0])
```

Заполните пропуски в значениях `mid_funding_at` на основании значений в столбцах `first_funding_at` и `last_funding_at`. В качестве нового значения вместо пропусков возьмите приблизительно середину интервала между этими двумя датами.

Оцените размер оставшихся пропусков в столбце.

```python
# Находим пропуски в mid_funding_at, которые мы сможем заполнить
mask = invest_df['mid_funding_at'].isna() & \
       invest_df['first_funding_at'].notna() & \
       invest_df['last_funding_at'].notna()

# Вычисляем середину интервала: к начальной дате прибавляем половину разницы
invest_df.loc[mask, 'mid_funding_at'] = invest_df.loc[mask, 'first_funding_at'] + \
    (invest_df.loc[mask, 'last_funding_at'] - invest_df.loc[mask, 'first_funding_at']) / 2

# Количество оставшихся пропусков
display(f"Осталось пропусков в mid_funding_at: {invest_df['mid_funding_at'].isna().sum()}")
```

Дальнейшие исследования выполняйте с учётом этой замены. Индивидуальные сегменты внутри средней и нишевой групп рассматривать не нужно — они объединяются в два общих сегмента.


```python
market_mapping = {}
for market, count in market_count.items():
    if count > 120:
        market_mapping[market] = market
    elif 35 <= count <= 120:
        market_mapping[market] = 'mid'  
    else:
        market_mapping[market] = 'niche'  

invest_df['market'] = invest_df['market'].map(market_mapping)

# Проверяем результат: топ-10 сегментов 
print("Обновленное распределение по сегментам:")
print(invest_df['market'].value_counts().head(10))
```

* Средние сегменты (mid) — это ниши, доказавшие жизнеспособность и готовые к масштабированию.

Роль сегментов:

* массовые рынки — «локомотивы» отрасли: обеспечивают стабильность, ликвидность и формируют стандарты;

* нишевые рынки — «лаборатория» инноваций: большинство проектов исчезнет, но лучшие станут новыми массовыми сегментами;

* средние сегменты — «социальный лифт» для перехода из нишевого статуса в мейнстрим.

**Итог: устойчивость массовых рынков дополняется инновационным потенциалом нишевых направлений — это основа сбалансированного развития отрасли.**

Определите компании с аномальным объёмом общего финансирования — используйте метод IQR отдельно по каждому сегменту. Напомним, что все нишевые сегменты должны быть объединены в одну группу, а средние — в другую.

Определите сегменты рынка с наибольшей долей компаний, получивших аномальное финансирование, и выведите топ таких сегментов.

```python
# Напишем функцию для определения аномалий 
def mark_outliers(group):
    Q1 = group['funding_total_usd'].quantile(0.25)
    Q3 = group['funding_total_usd'].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    group['is_outlier'] = group['funding_total_usd'] > upper_bound
    return group

invest_df = invest_df.groupby('market', group_keys=False).apply(mark_outliers)
print(f"Всего выявлено аномальных компаний: {invest_df['is_outlier'].sum()}")
```

```python
# Выводим Топ-10 в процентах
outlier_shares = invest_df.groupby('market')['is_outlier'].mean().sort_values(ascending=False)
outlier_shares_pct = round((outlier_shares * 100), 2)
print("Топ сегментов по доле компаний с аномальным финансированием (%):")
print(outlier_shares_pct.head(10))
```

В некоторых отраслях инвестиции распределяются очень неравномерно: несколько стартапов получают намного больше денег, чем остальные. ### 4.3 Годовая динамика доли возвращённых средств по типам финансирования

Заказчик хочет знать, какая часть вложенных или выданных денег со временем возвращается обратно инвесторам или финансистам. Ваша цель — для каждого года и каждого вида финансирования рассчитать нормированные значения возврата средств: то есть какую долю возвращённые средства составляют от предоставленных. При этом слишком большие аномальные значения, то есть неадекватные выбросы, нужно заменить на пропуски.

Совет: когда будете делить сумму возвращённых средств на суммарный объём привлечённого финансирования по конкретному году, добавьте к знаменателю небольшое число, например `1e-60`. Это поможет избежать деления на ноль.

```python
# Агрегируем инвестиции и приводим к млн (делим на 1 000 000)
funding_types = ['seed', 'venture', 'equity_crowdfunding', 'undisclosed', 
                 'convertible_note', 'debt_financing', 'angel', 'grant', 
                 'private_equity', 'post_ipo_equity', 'post_ipo_debt', 
                 'secondary_market', 'product_crowdfunding']

invest_by_year_mln = invest_df_final.groupby(invest_df_final['mid_funding_at'].dt.year)[funding_types].sum() / 1_000_000

# Находим общие годы в интервале 2000-2014
common_years = invest_by_year_mln.index.intersection(returns_df.index)
common_years = [y for y in common_years if 2000 <= y <= 2014]

# Рассчитываем реальную долю возврата (mln / mln) в процентах
norm_returns_correct = round((returns_df.loc[common_years, funding_types] / (invest_by_year_mln.loc[common_years] + 1e-60) * 100), 2)

# Очистка от аномалий
def replace_outliers_with_nan(df):
    df_clean = df.copy()
    for col in df_clean.columns:
        Q1, Q3 = df_clean[col].quantile(0.25), df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        df_clean.loc[df_clean[col] > (Q3 + 1.5 * IQR), col] = np.nan
    return df_clean

norm_returns_final = replace_outliers_with_nan(norm_returns_correct)

display(norm_returns_final)
```

Постройте график, на котором отобразите нормированные значения возврата средств для типов финансирования `venture`, `debt_financing`, `private_equity`, `seed` и `angel`.

Сделайте вывод о том, в каких типах финансирования наблюдается наиболее устойчивый рост показателя.

```python
# Визуализация выбранных типов
selected_types = ['venture', 'debt_financing', 'private_equity', 'seed', 'angel']
plt.figure(figsize=(15, 8))
sns.lineplot(data=norm_returns_final[selected_types], dashes=False, marker='o', linewidth=2)

# Добавляем линию окупаемости 100%
plt.axhline(y=100, color='red', linestyle='--', alpha=0.5, label='Порог окупаемости (100%)')

plt.title('Динамика доли возврата средств', fontsize=15)
plt.ylabel('Процент возврата')
plt.xlabel('Год инвестирования')
plt.grid(True, linestyle='--', alpha=0.4)
plt.xlim(2000, 2014)
plt.legend(title='Тип финансирования', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Самый устойчивый рост значения возврата средств показыва мы можем увидеть у типа `venture`. С 2009 года значения только увеличиваются. Также можно отметить тип `debt_finansing`, у которого также с 2009 года происходило увеличение доли возврата, однако в 2012-2013 годах рост остановился и даже немного откатился. После 2013 возврат средств уже показал рост.

В остальных категориях устойчивого роста не наблюдается. Этапы роста чередуются со спадами.
