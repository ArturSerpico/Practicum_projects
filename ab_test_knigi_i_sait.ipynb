## 2. Проверка гипотезы в Python

Гипотеза звучит так: пользователи из Санкт-Петербурга проводят в среднем больше времени за чтением и прослушиванием книг в приложении, чем пользователи из Москвы. Попробуйте статистически это доказать, используя одностороннюю проверку гипотезы с двумя выборками:

- Нулевая гипотеза H₀: Средняя активность пользователей в часах в двух группах (Москва и Санкт-Петербург) не различается.

- Альтернативная гипотеза H₁: Средняя активность пользователей в Санкт-Петербурге больше, и это различие статистически значимо.

Так как необходимо сравнить среднее время пользователя в приложении, для проверки гипотезы следует использовать ***t-тест***. Проверять гипотезу будем с уровнем значимости 5%.

```python
# сохраняем выборки
msk = df_knigi[df_knigi.city == 'Москва'].hours
spb = df_knigi[df_knigi.city == 'Санкт-Петербург'].hours

# устанавливаем уровень значимости
alpha = 0.05

# проводим t-тест
stat_ttest, p_value_ttest = ttest_ind(
    msk, 
    spb, 
    alternative='less'
) 

# выводим полученное p-value 
print(f'pvalue={p_value_ttest}') 

if p_value_ttest > alpha:
    print('Нулевая гипотеза находит подтверждение')
else:
    print('Нулевая гипотеза не находит подтверждения')
```

<div class="alert alert-danger">

**❌ Комментарий ревьюера ❌**
   
   Касательно теста Манна–Уитни: он действительно подходит для сравнения распределений в целом или медиан при непараметрических данных. Однако в нашей задаче метрика — среднее время пользователя в приложении — предполагает сравнение средних значений, а не медиан. Поэтому правильнее использовать t-тест - его можно применять при ненормальном распределении данных, потому что он основан не на самих исходных значениях, а на распределении выборочного среднего. Согласно центральной предельной теореме, при достаточно большом объёме выборки распределение среднего стремится к нормальному независимо от формы исходного распределения (при условии независимости наблюдений и конечной дисперсии). Поэтому t-статистика, которая строится из среднего и его стандартной ошибки, также приближается к t-распределению, что и позволяет проводить тест даже при ненормальных данных.

## 3. По таблице `ab_test_participants` оцените корректность проведения теста:

   3\.1 Выделите пользователей, участвующих в тесте, и проверьте:

   - соответствие требованиям технического задания,

   - равномерность распределения пользователей по группам теста,

   - отсутствие пересечений с конкурирующим тестом (нет пользователей, участвующих одновременно в двух тестовых группах).

```python
# проверим пересечение с конкурирующим тестом
interface_users = participants[participants.ab_test == 'interface_eu_test']
recommender_users = participants[participants.ab_test == 'recommender_system_test']

common_users = pd.merge(interface_users, recommender_users, on='user_id')

print(f'Количество пользователей, которые учавствовали одновременно в двух A/B тестах: {len(common_users)}')
```

887 пользователей одновременно участвовало в двух A/B тестах. Исключим этих пользователей.

```python
# Находим user_id, у которых встречается только interface_eu_test (нет recommender_system_test)
users_only_interface = (
    participants
    .groupby('user_id')['ab_test']
    .apply(lambda x: set(x) == {'interface_eu_test'})
    .loc[lambda s: s].index
)

# Фильтруем исходный датафрейм: оставляем только этих пользователей и только тест interface_eu_test
interface_test = (
    participants[participants['user_id'].isin(users_only_interface)]
    .query("ab_test == 'interface_eu_test'")
    .reset_index(drop=True)
)

# Проверим есть ли пользователи, вошедшие в обе группы
print(f'Пользователей, вошедших в обе группы', interface_test.duplicated(subset=['user_id']).sum())

# Проверим равномерность распределения пользователей в группах
print('\nКоличество пользователей в группах\n', interface_test.groupby('group')['user_id'].count())
```

- оставьте только события, связанные с участвующими в изучаемом тесте пользователями;

```python
# Оставляем в ab_test_events только события пользователей, участвующих в изучаемом тесте
interface_test_events = events[events['user_id'].isin(interface_test['user_id'].unique())].copy()

# Сбрасываем индекс
interface_test_events = interface_test_events.reset_index(drop=True)
```

- определите горизонт анализа: рассчитайте время (лайфтайм) совершения события пользователем после регистрации и оставьте только те события, которые были выполнены в течение первых семи дней с момента регистрации;

```python
# Находим дату регистрации для каждого user_id
registration_dates = (
    interface_test_events[interface_test_events['event_name'] == 'registration']
    .groupby('user_id')['event_dt']
    .first()
    .rename('registration_dt')
)

# Присоединяем дату регистрации к основному датафрейму
interface_test_events = interface_test_events.merge(
    registration_dates,
    left_on='user_id',
    right_index=True,
    how='left'
)

# Оставляем только события в первые 7 дней с момента регистрации
seven_days = pd.Timedelta(days=7)
interface_test_events_7d = interface_test_events[
    (interface_test_events['event_dt'] >= interface_test_events['registration_dt']) &
    (interface_test_events['event_dt'] < interface_test_events['registration_dt'] + seven_days)
].copy()

# Убираем вспомогательный столбец
interface_test_events_7d = interface_test_events_7d.drop(columns=['registration_dt'], errors='ignore')

# Сбрасываем индекс
interface_test_events_7d = interface_test_events_7d.reset_index(drop=True)
```

Оцените достаточность выборки для получения статистически значимых результатов A/B-теста. Заданные параметры:

- базовый показатель конверсии — 30%,

- мощность теста — 80%,

- достоверность теста — 95%.

```python
# Задаём параметры
alpha = 0.05
power = 0.8
p = 0.3
mde = 0.03
effect_size = proportion_effectsize(p, p + mde)

# Инициализируем класс NormalIndPower
power_analysis = NormalIndPower()

# Расчёт размера выборки
sample_size = power_analysis.solve_power(
    effect_size = effect_size,
    power = power,
    alpha = alpha,
    ratio = 1
)

print(f"Необходимый размер выборки для каждой группы: {int(sample_size)}")

if int(sample_size) <= len(interface_test[interface_test.group == 'A']) and int(sample_size) <= len(interface_test[interface_test.group == 'B']):
    print('Размеры контрольной и тестовой выборок достаточны')
else:
    print('Размеры контрольной и тестовой выборок не достаточны')
```

```python
# посчитаем размер группы А
n_a = events_merged_test[events_merged_test.group == 'A']['user_id'].nunique()

# посчитаем размер группы B
n_b = events_merged_test[events_merged_test.group == 'B']['user_id'].nunique()

# посчитаем количество посетителей из группы А, сделавших покупку
m_a = events_merged_test[(events_merged_test.event_name == 'purchase') & (events_merged_test.group == 'A')]['user_id'].nunique()

# посчитаем количество посетителей из группы B, сделавших покупку
m_b = events_merged_test[(events_merged_test.event_name == 'purchase') & (events_merged_test.group == 'B')]['user_id'].nunique()

# рассчитаем долю посетителей, совершивших покупку для выборки А
p_a = m_a / n_a

# рассчитаем долю посетителей, совершивших покупку для выборки B
p_b = m_b / n_b

print(f'Доля покупок в первые семь дней с момента регистрации пользователей контрольной группы составила {p_a:.2%}')
print(f'Доля покупок в первые семь дней с момента регистрации пользователей тестовой группы составила {p_b:.2%}')
print(f'Разница между показателями: B - A = {(p_b - p_a):.2%}')
```

## 4. Проведите оценку результатов A/B-тестирования:

- Проверьте изменение конверсии подходящим статистическим тестом, учитывая все этапы проверки гипотез.

Гипотеза в данном исследовании заключается в следующем: упрощение интерфейса приведёт к тому, что в течение семи дней после регистрации в системе конверсия зарегистрированных пользователей в покупателей увеличится как минимум на три процентных пункта. Таким образом, были сформулированы гипотезы:
* Нулевая гипотеза H0: Количество пользователей совершающих покупку в новой версии сайта и количество пользователей совершающих покупку в старой версии сайта не различается.
* Альтернативная гипотеза H1: Количество пользователей совершающих покупку в новой версии сайта больше, чем количество пользователей совершающих покупку в старой версии сайта.
